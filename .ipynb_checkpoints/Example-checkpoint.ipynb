A simple example showing how to draw a light curve using the templates

```python
%matplotlib inline
import numpy as np
print(np.__version__)
import matplotlib.pylab as plt
import matplotlib.gridspec as gridspec
import pickle

# Loading ab-type RR Lyrae templates
with open('templates/RRab_LSST.pkl', "rb") as f:
    lc_data, lc_info, lc_per = pickle.load(f, encoding='latin1')

obj_id = sorted(list(lc_data.keys()))
filters = [b'u', b'g', b'r', b'i', b'z']
```

```python
idx = 0  # light curve index
N = 24  # number of samples
rng = np.random.RandomState(0)
white_noise = rng.randn(N, len(filters))
data = lc_data[obj_id[idx]]
period = lc_per[obj_id[idx]]

fig = plt.figure(figsize=(8, 12))
gs = gridspec.GridSpec(len(filters), 2)

for i, filt in enumerate(filters):
    mjd = data[filt]['mjd'][:N]
    mag = data[filt]['mag'][:N]                    
    err = data[filt]['error'][:N]
    # Add N(0, err**2) to the template mags
    noisy_mag = mag + err*white_noise[:N, i]
    phi = np.mod(mjd, period)/period
    ax = plt.subplot(gs[i, 0])
    ax.invert_yaxis()
    ax.errorbar(mjd, noisy_mag, err, fmt='.')
    ax.grid()
    ax.set_ylabel(filt.decode('UTF-8')+" band magnitude")
    ax.set_xlabel("MJD")
    ax = plt.subplot(gs[i, 1])
    ax.invert_yaxis()
    I = np.argsort(phi)    
    ax.errorbar(np.hstack((phi, phi+1)), 
                np.hstack((noisy_mag, noisy_mag)), 
                np.hstack((err, err)), label='noisy', fmt='.')
    ax.plot(np.hstack((phi[I], phi[I]+1)), 
                np.hstack((mag[I], mag[I])), linewidth=4, label='underlying', alpha=0.5)
    ax.set_xlabel("Phase")
    if i == 0:
        plt.legend(loc=4)
    ax.grid()
gs.tight_layout(fig)        
```
